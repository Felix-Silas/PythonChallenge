```python
import random

height, weight = [], []

for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    
plt.rc('font', family='Malgun Gothic')
plt.title('  ')
plt.scatter(height,weight, c=height, cmap='RdPu') #c         ,   cmap 
plt.xlabel('')
plt.ylabel('')
plt.colorbar(label='')
plt.show()
```

```python
import random

height, weight, score = [], [], []

for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    score.append(random.randint(1,1000))
    
    
plt.rc('font', family='Malgun Gothic')
plt.title(' ,   ')
plt.scatter(height,weight, s=score, c=score, cmap='rainbow', alpha = 0.7) #x, y,      ,   s  
plt.xlabel('')
plt.ylabel('')
plt.colorbar(label=' ')
plt.show()
```
