```python
import random

print('구구단을 외자! 구구단을 외자!')
for i in range(10):
    a = random.randint(2,9)
    b = random.randint(1,9)
    answer = input(str(a) + '*' + str(b) + '? ')
    answer = int(answer)
    if answer != a*b:
        print('땡! 정답은', a*b)
        break
```

```python
import random
print('구구단을 외자! 구구단을 외자!')
answer, a, b = 0, 0, 0
while answer == a * b:
    a = random.randint(2,9)
    b = random.randint(1,9)
    answer = input(str(a) + '*' + str(b) + '? ')
    answer = int(answer)
print('땡! 정답은', a*b)
```
