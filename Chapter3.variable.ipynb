```python
p = 3.141592
r = 5
result = p * r ** 2
print('원의 넓이 =', p * r ** 2)
print(result)
```

```python
"""
원의 넓이를 구하는 프로그램입니다. 
변수 p는 원주율을 저장한 변수이고, 
변수 r는 반지름을 저장한 변수입니다.
"""

p = 3.141592
r = 5

#원의 넓이
result1 = p * r ** 2
print('원의 넓이 =', result1)

#원의 둘레
result2 = 2 * p * r
print('원의 둘레 =', result2)

#구의 겉넓이
result3 = 4 * p * r ** 2
print('구의 겉넓이 =', result3)
```

```python
p=3.141592
r = int(input('반지름은? '))

result1 = p * r ** 2
result2 = 2 * p * r
result3 = 4 * p * r ** 2

print('원의 넓이 =', result1, '원의 둘레 =', result2, '구의 겉넓이 =', result3)
```

```python
p=3.141592
r = int(input('반지름은?'))

c_area = p * r ** 2
c_len = 2 * p * r
s_area = 4 * p * r ** 2

print('원의 넓이 =', c_area, '원의 둘레 =', c_len, '구의 겉넓이 =', s_area)
```
