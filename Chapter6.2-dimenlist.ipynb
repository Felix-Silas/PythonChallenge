```python
# 순위 국가 금메달 은메달 동메달 합계
olympic = [[1, '미국', 46, 37, 38 , 121],
           [2, '영국', 27, 23, 17, 67],
           [3, '중국', 26, 18, 26, 70],
           [4, '러시아', 19, 18, 19, 55],
           [5, '독일', 17, 10, 15, 42],
           [6, '일본', 12, 8, 21, 41],
           [7, '프랑스', 10, 18, 14, 42],
           [8, '대한민국', 9, 3, 9, 21]]

print(olympic[0][1], '금메달 개수:', olympic[0][2])
print(olympic[4][1], '금메달 개수:', olympic[4][2])
```

```python
print(olympic[0][1], '금은동 개수:', olympic[0][2:5])
print(olympic[4][1], '금은동 개수:', olympic[4][2:5])
print(olympic[:4])
print(olympic[:4][1])
```

```python
for row in olympic:
    print(row[:2])
    print(row[:2][1])
```

```python
print('금메달보다 은메달이 많은 나라')
for row in olympic:
    if row[2] < row[3]:
        print(row[1])
        
print('금메달보다 동메달이 많은 나라')
for row in olympic:
    if row[2] < row[4]:
        print(row[1])
```

```python
total_gold = 0

for row in olympic:
    total_gold += row[2]

print(total_gold)
```

```python
for row in olympic:
    gold_per = row[2] / row[5] * 100
    if gold_per > 40:
        print(row[1], gold_per, '%')
```
