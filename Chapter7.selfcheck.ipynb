```python
day = {'월요일': 'Mon', '화요일': 'Tue', '수요일': 'Wed', '목요일': 'Thu', '금요일': 'Fri'}
day['토요일'] = 'Sat'
day['일요일'] = 'Sun'
print(day)
```

```python
del day['수요일']
print(day)
```

```python
book = {'역사대모험': 20000, '영단어': 9000, '파이썬': 17000, '여행에세이': 22000, '삼국지': 33000}

for v, p in book.items():
    if p > 15000:
        book[v] *= 1.05
    else:
        book[v] *= 1.1
print(book)
```
