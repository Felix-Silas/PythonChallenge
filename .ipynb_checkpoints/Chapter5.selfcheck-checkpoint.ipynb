```python
print('홈트레이닝 2주 계획입니다.')

for i in range(1,15):
    if i % 5 == 0:
        print(i, '일: 휴식일입니다.')
    else:
        print(i, '일: 스쿼트 - 윗몸 일으키기 - 유산소 운동')
```
