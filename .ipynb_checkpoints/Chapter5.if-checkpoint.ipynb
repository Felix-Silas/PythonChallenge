```python
import random

print('구구단을 외자! 구구단을 외자!')

for i in range(10):
    a = random.randint(2,9)
    b = random.randint(1,9)
    question = str(a) + '*' + str(b) + '? '
    answer = input(question)
    answer = int(answer)
    if answer != a * b:
        print('틀렸습니다.')
```

```python
import random

a = random.randint(1,30)

if a % 2 == 0:
    print(a)
    print('짝수')
    
if a % 2 == 1:
    print(a)
    print('홀수')
```

```python
for i in range(1,31):
    if i % 2 == 0:
        print(i, '짝수', end = ', ')
    
    else:
        print(i, '홀수', end = ', ')
```

```python
import random

print('구구단을 외자! 구구단을 외자!')

for i in range(10):
    a = random.randint(2,9)
    b = random.randint(1,9)
    question = str(a) + '*' + str(b) + '? '
    answer = input(question)
    answer = int(answer)
    if answer != a * b:
        print('틀렸습니다. 5번 반복해 읽어 봅시다.')
        for j in range(5):
            print(a, '*', b, '=', a*b)
```
