```python
midterm = {'도윤': 43, '하윤': 82, '시우': 76, '지유': 61, '주원': 94}

student = input('어떤 학생의 점수가 궁금한가요? ')

if student in midterm.keys():
    print(midterm[student], '점입니다.')

else:
    print('해당 학생이 없습니다.')
    
```
