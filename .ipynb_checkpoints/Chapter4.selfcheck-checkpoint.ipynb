```python
cold_yn = input('기침을 하십니까? (Y/N) ')
temperature = float(input('체온은 몇 도입니까? '))

if cold_yn == 'Y' and temperature >= 38.5:
    print('독감일 수 있습니다.')
```

```python
food = input('음식을 입력해주세요. (라면/김밥) ')

if food == '라면':
    print('기본라면(3,500원), 떡라면(4,000원), 만두라면(4,000원)')
elif food == '김밥':
    print('야채김밥(2,500원), 참치김밥(3,500원)')
else:
    print('올바른 메뉴를 입력해 주세요.')
    
```

```python
score = int(input('점수를 입력해주세요. '))

if score > 80 and score <= 100:
    print('A')
elif score >= 40 and score < 80:
    print('B')
elif score < 40 and score >= 0:
    print('C')
else:
    print('잘못된 숫자를 입력했습니다.')
```

```python
score = int(input('점수를 입력해주세요. '))

if score > 100 or score < 0:
    print('잘못된 숫자를 입력했습니다.')
elif score >= 80:
    print('A')
elif score >= 40:
    print('B')
else:
    print('C')
```
