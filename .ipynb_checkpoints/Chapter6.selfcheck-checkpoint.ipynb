```python
import random

book_name = ['역사탐험대', '파이썬', '학습법', '영단어', '여행에세이', '삼국지']
book_price = []

for i in range(len(book_name)):
    book_price.append(random.randint(15000,25000))
print(book_name)
print(book_price)

expensive = max(book_price)
idx = book_price.index(expensive)
print('가장 비싼 책은', book_name[idx])
```

```python
ranking =[[1, '라이언', 956412, 564553],
         [2, '니니즈', 861832, 927175],
         [3, '어피치', 796354, 593175],
         [4, '프로도', 534840, 864326],
         [5, '네오', 387896, 648367]]

print('참가자의 순위-이름')
for rank in ranking:
    print(rank[:2])

print('이번 주 득표수가 50만 표가 넘는 참가자')
for rank in ranking:
    if rank[2] > 500000:
        print(rank[1:3])

print('지난주보다 득표수가 떨어진 참가자')
for rank in ranking:
    if rank[2] < rank[3]:
        print(rank[1], rank[3] - rank[2], '표 하락')

print('지난주와 이번 주 득표수 중 한 번이라도 80만 표 넘은 참가자')
for rank in ranking:
    if rank[2] > 800000 or rank[3] > 800000:
        print(rank[1], '80만 표 돌파')
```
