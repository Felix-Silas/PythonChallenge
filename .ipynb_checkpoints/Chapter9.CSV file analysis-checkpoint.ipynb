```python
import csv  # csv 라이브러리 포함하기

f = open('card.csv', encoding='utf8')   # 파일 열기  
data = csv.reader(f)   # 파일 읽기
next(data)   # 표의 헤더 제거하기
data = list(data) # 리스토로 변환하기

print(data)
```

```python
spend = 0

for row in data:
    if row[-1] == '전표매입':
        spend += int(row[6])
    
print(spend)
```

```python
import matplotlib.pyplot as plt

s_mon = [0, 0, 0]

for row in data:
    if row[-1] == '전표매입':
        mon = int(row[0].split('-')[1])
        payment = int(row[6])
        idx = mon - 10
        s_mon[idx] += payment
print(s_mon)

plt.rc('font', family = 'Malgun Gothic')
plt.title('10~12월 지출현황')
plt.bar(['10월', '11월', '12월'], s_mon, color = 'royalblue')
plt.show()
```

```python
taxi_mon = [0, 0, 0]

for row in data:
    if row[-1] == '전표매입' and '택시' in row[5]:
        mon = int(row[0].split('-')[1])
        payment = int(row[6])
        idx = mon - 10
        taxi_mon[idx] += payment

plt.rc('font', family = 'Malgun Gothic')
plt.title('10~12월 택시비 지출현황')
plt.plot(['10월', '11월', '12월'], taxi_mon, color = 'crimson', label = '택시비 지출액')
plt.legend()
plt.show()
```

```python
deli = [0, 0, 0]

for row in data:
    if row[-1] == '전표매입' and row[5] == '(주)우아한형제들':
        mon = int(row[0].split('-')[1])
        payment = int(row[6])
        idx = mon - 10
        deli[idx] += payment

#plt.rc('font', family = 'Malgun Gothic')
plt.title('10~12월 배달음식비 지출현황')
plt.plot(['10월', '11월', '12월'], deli, color = 'indigo', label = '배달음식비 지출액')
plt.legend()
plt.show()
```

```python
plt.title('10~12월 택시비/배달음식비 지출현황')
plt.plot(['10월', '11월', '12월'], taxi_mon, color = 'crimson', label = '택시비 지출액')
plt.plot(['10월', '11월', '12월'], deli, color = 'indigo', label = '배달음식비 지출액')
plt.legend()
plt.show()
```

```python
import operator
import matplotlib.pyplot as plt

spending = {}

for row in data:
    if row[-1] == '전표매입':
        store = row[-4]
        payment = int(row[-3])
        
        if store not in spending.keys():
            spending[store] = payment
            
        else:
            spending[store] += payment

top10 = sorted(spending.items(), key=operator.itemgetter(1), reverse = True)[:10]
top10_store, top10_amount = [], []

for t in top10:
    top10_store.append(t[0])
    top10_amount.append(t[1])

plt.rc('font', family = 'Malgun Gothic')
plt.title('10월~12월 지추ㅜㄹ Top 10')
plt.barh(top10_store, top10_amount, color = 'b')
plt.show()
```
