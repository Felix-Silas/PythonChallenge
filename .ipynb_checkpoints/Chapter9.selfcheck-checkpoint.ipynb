```python
import csv

f = open('card.csv', encoding='utf8')
data = csv.reader(f)
next(data)
data = list(data)
print(data)
```

```python
payment_cancle = 0


for row in data:
    if row[-1] != '전표매입':
        payment_cancle += int(row[-3])

print(payment_cancle)
```

```python
import matplotlib.pyplot as plt

store_mon = [0, 0, 0]

for row in data:
    if row[-1] == '전표매입' and '백화점' in row[-4]:
        mon = int(row[0].split('-')[1])
        idx = mon - 10
        store_mon[idx] += int(row[-3])

plt.rc('font', family = 'Malgun Gothic')
plt.title('10~12월 백화점 지출액')
plt.plot(['10월', '11월', '12월'], store_mon)
plt.show()
```
