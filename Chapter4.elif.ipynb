```python
age = int(input('나이가 어떻게 되세요? '))
if age >= 20:
    print('성인은 15,000원입니다.')
    price = 15000;
else:
    print('미성년자는 6,000원입니다.')
    price = 6000;

price_input = int(input('입장료를 넣어 주세요. --> '))
if price_input > price:
    print(price_input - price, '원을 거슬러 드리겠습니다.')
elif price_input == price:
    print('감사합니다. 즐거운 시간 되세요!')
else:
    print(price - price_input, '원을 더 내야 합니다.')
```

```python
price = int(input('구매금액은? '))
if price < 20000:
    print('주문금액이 부족합니다.')
elif price < 50000:
    print('배송비 2500원이 추가됩니다.')
    price += 2500
else:
    print('무료배송됩니다.')
print('최종 결제 금액은', price, '원입니다.')
```
