```python
import random

score = []
for i in range(25):
    score.append(random.randint(0,100))
    print(score)
    
for s in score:
    if s >= 70:
        print(score.index(s) + 1, '번:', s)

for i in range(len(score)):
    if score[i] >= 70:
        print(i+1, '번:', score[i])
```
