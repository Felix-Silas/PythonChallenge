```python
snack = int(input('과자를 몇 개 구매하겠습니까?'))
ramen = int(input('라면을 몇 개 구매하겠습니까?'))
ham = int(input('과자를 몇 개 구매하겠습니까?'))

price = snack * 1200 + ramen * 800 + ham * 2400


print('할인 전 금액:', price, ', 할인 후 금액:', price * 0.75 )
```

```python
num_1 = int(input('숫자1 입력: '))
num_2 = int(input('숫자2 입력: '))

print('두 수를 더한 값:', num_1 + num_2)
print('두 수를 뺀 값:', num_1 - num_2)
print('두 수를 곱한 값:', num_1 * num_2)
print('두 수를 나눈 값:', num_1 / num_2)
```
