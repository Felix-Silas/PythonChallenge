```python
import folium

name = '해운대 해수욕장'
lat = '35.15870'
long = '129.16047'

map = folium.Map([lat,long], zoom_start = 11)

folium.Marker([lat,long], tooltip = name, icon = folium.Icon(color = 'red', icon = 'hand-o-down', prefix ='fa')).add_to(map)
map
```

```python
import folium

name = ['해운대 해수욕장', '부산대학교', '김해국제공항']
lat = [35.15870, 35.23360, 35.17357]
long = [129.16047, 129.08309, 128.94632]
icons = ['hand-o-down', 'university', 'plane']
colors = ['red', 'blue', 'black']

map = folium.Map([lat[0],long[0]], zoom_start = 11)

for i in range(len(name)):
    folium.Marker([lat[i],long[i]], tooltip = name[i], 
                  icon = folium.Icon(color = colors[i], icon = icons[i], prefix ='fa')).add_to(map)
map
```
